# 综合题(选做)

任务要求：白葡萄酒质量共分为7个等级，对白葡萄酒质量进行预测。
1. 使用精度作为评价指标。
2. 使用一元线性回归、多元线性回归、对数线性回归等线性回归模型对葡萄酒质量进行预测，计算其十折交叉验证的精度。
3. 使用对数几率回归、线性判别分析对葡萄酒质量进行预测，计算其十折交叉验证的精度。
4. 对比不同模型的预测能力。

## 读取数据

```python
import pandas as pd
import numpy as np
data = pd.read_csv('data/wine_quality/winequality-white.csv', delimiter=";")
```

```python
data.head()
```

```python
# YOUR CODE HERE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_predict
from sklearn.linear_model import LinearRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import math

datax = data.iloc[:,:11]
datay = data.iloc[:,11:]

data1x = datax.iloc[:,0:1]

model1 = LinearRegression()
model1_prediction = cross_val_predict(model1,data1x,datay,cv=10)
print("一元线性回归：")
model1_prediction = np.rint(model1_prediction)
print(accuracy_score(datay.values.flatten(),pd.DataFrame(model1_prediction).values.flatten()))

model2_prediction = cross_val_predict(model1,datax,datay,cv=10)
print("多元线性回归：")
model2_prediction = np.rint(model2_prediction)
print(accuracy_score(datay.values.flatten(),pd.DataFrame(model2_prediction).values.flatten()))

data1y = pd.DataFrame(np.log(np.array(datay)))
model3_prediction = cross_val_predict(model1,datax,data1y,cv=10)
print("对数线性回归：")
model3_prediction = np.rint(np.exp(model3_prediction))
print(accuracy_score(datay.values.flatten(),pd.DataFrame(model3_prediction).values.flatten()))

model4 = LogisticRegression()
model4_prediction = cross_val_predict(model4,datax,datay,cv=10)
print("对数几率回归：")
model4_prediction = np.rint(model4_prediction)
print(accuracy_score(datay.values.flatten(),pd.DataFrame(model4_prediction).values.flatten()))

model5 = LinearDiscriminantAnalysis()
model5_prediction = cross_val_predict(model5,datax,datay,cv=10)
print("线性判别分析：")
model5_prediction = np.rint(model5_prediction)
print(accuracy_score(datay.values.flatten(),pd.DataFrame(model5_prediction).values.flatten()))
```
